```python
import cv2
import os
from keras.preprocessing.image import img_to_array
```

```python
data=[]
for imagePath in os.listdir('pokemon_imgs/'):
    # load the image, pre-process it, and store it in the data list
    image = cv2.imread('pokemon_imgs/' + imagePath)
    image = cv2.resize(image, (64, 64))
    image = img_to_array(image)
    data.append(image)
    #print(imagePath)
```

```python
len(data)
```

```python
import numpy as np
```

```python
data = np.array(data)
```

```python
dcgan_data = dcgan_data.DCGAN()
```

```python
dcgan_data.train(data, epochs=4000, save_interval=10)
```
